```python
import pandas as pd
import numpy as np

#Problem statement1, pat1: read the dataset Laliga.csv
#skip rows = 1 included as there is an empty line at the top which makes the column names as unnamed if not skipped
Laliga = pd.read_csv("Laliga.csv", skiprows=1)
Laliga.info()
```

```python
#Sanity check on the data loaded
Laliga.head(10)
Laliga.tail(10)
```

```python
#Problem Statement1, pat2: Replace dashed data with 0
#Ignore Pos, Team, Debut, BestPosition columns as they do not have '-'
#Ignore In Debut and Since/LastApp has '-' in a different context(as a date format)
Laliga.iloc[:,3:17] = Laliga.iloc[:,3:17].replace('-',0)
#check if the '-'s are replaced with 0s 
print(Laliga.head(10))
print(Laliga.tail(10))

#Convert all Numeric data to numeric data type after replacing '-' to 0 in order to perform arithmetic operations
Laliga = Laliga.apply(pd.to_numeric,errors='ignore')
#Check if the data types are changed
Laliga.info()
```

```python
#Problem Statement2: To print all teams which started playing between 1930-1980 (using Debut column)
Laliga['Debut_year1'] = Laliga['Debut'].str.slice(0,4)
Laliga.Debut_year1 = Laliga.Debut_year1.astype('int64')
print("Teams that started playing between 1930 and 1980: ")
print(Laliga.loc[(Laliga['Debut_year1']>1930)
                 &(Laliga['Debut_year1']<1980)]['Team'])
#Debut year1 column is no longer required so dropping it after use
Laliga.drop(['Debut_year1'], axis=1)
```

```python
#Problem Statement3: List of teams which came Top 5 in terms of points
print("Top 5 teams in terms of points:")
print(Laliga.sort_values(by='Points', ascending=False)['Team'].head(5))
```

```python
#Problem Statement4: Define Goal difference column and use the same to fine the teams with maximum and minimum Goal Difference 
def Goal_diff_count(row):
    return abs(row['GoalsFor'] - row['GoalsAgainst']), row['Team']

#Since the ask in the requirement is team with max goal difference rather than the maximum goal difference value itself, only team is printed
print("Team with maximum goal difference: ", 
      Laliga.apply(Goal_diff_count, axis=1).max()[1])
#For minimum Number of Games Played is validated because if that is not done, the minimum will incorrectly point to the team Girona which has GoalsFor and GoalsAgainst as 0. Ideally it should be ignore as they dont qualify for Goal difference check
print("Team with minimum goal difference: ",
      Laliga.loc[Laliga['GamesPlayed']>0].apply(Goal_diff_count, axis=1).min()[1])
```

```python
#Problem statement5: Add column Winning_Percent to the dataset
#fillna(0) is used to handle divide by 0 exception for Ex, Gorina
Laliga['Winning_Percent'] = ((Laliga['GamesWon']/Laliga['GamesPlayed'])*100).fillna(0)
print(Laliga.sort_values(by='Winning_Percent', ascending = False)['Team'].head(5))
```

```python
#Problem Statement6: Group team by Best Position and print the sum of their points for all positions
BestPosition_grp = Laliga.groupby(['BestPosition'])
print(BestPosition_grp['Points'].agg(np.sum))
```
